```python
split=[0.6,0.7,0.8,0.9]
Split_Percent=[]
Score=[]
Coef=[]
Intercept=[]
MSE=[]
R2_score=[]

```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
for i in range(len(split)):
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=split[i], random_state=42)
    reg = LinearRegression().fit(X_train, y_train)
    print(split[i]*100,':',100-split[i]*100)
    a=split[i]*100
    b=100-split[i]*100
    Split_Percent.append(str(a)+'-'+str(b))
    print('score:',reg.score(X_train, y_train))
    Score.append(reg.score(X_train, y_train))
    print('coef:',reg.coef_)
    Coef.append((reg.coef_).tolist())
    print('intercept: ',reg.intercept_)
    Intercept.append(reg.intercept_)
    y_pred=reg.predict(X_test)
    print('mse: ',mean_squared_error(y_pred,y_test))
    MSE.append(mean_squared_error(y_pred,y_test))
    print(r2_score(y_test,y_pred))
    R2_score.append(r2_score(y_test,y_pred))
    print('')   
```

```python
Coef_arr=np.array(Coef)

df = pd.DataFrame(list(zip(Split_Percent,Score,Intercept,Coef_arr[:,0],Coef_arr[:,1],Coef_arr[:,2],Coef_arr[:,3],Coef_arr[:,4],Coef_arr[:,5],MSE,R2_score )),
               columns =['Split_Percent','Score','Intercept','coef_1','coef_2','coef_3','coef_4','coef_5','coef_6','MSE','R2_score'])
df
# split=[0.6,0.7,0.8,0.9]
# Split_Percent=[]
# Score=[]
# Coef=[]
# Intercept=[]
# MSE=[]
# R2_score=[]
```

```python
data = pd.read_csv("haberman.data",sep=',')
data
```

```python
feature=data.iloc[:,:3]
feature.head()
```

```python
label=data.iloc[:,-1]
label.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature,label, train_size=0.80, random_state=42)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=42).fit(X_train, y_train)
y_pred=clf.predict(X_test)
print(y_pred.tolist())
print('score: ',clf.score(X_train, y_train))
print('coef: ',clf.coef_)
print('intercept: ', clf.intercept_)
print(y_test.tolist())
```

```python
def log_reg_error(yp,yt):
    sum=0
    for i in range(len(yp)):
        if yp[i]!=yt[i]:
            sum+=1
    return sum/len(yp)
```

```python
print('error: ',log_reg_error(y_pred.tolist(),y_test.tolist()))
```
